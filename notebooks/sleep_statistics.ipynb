```python
import os
import pandas as pd
import yasa
import datetime
```

```python
sleep_data = pd.read_csv(os.path.join('..', 'data','pmi-mini_pilot','4de37ee3-7baf-4592-a431-a6d01fff6bd1',
                                      'WRSD-PMI-01_c3ea8756-8e12-45b7-a31c-1a03e398daf9',
                                      'garmin-connect-sleep-stage',
                                      '000000_garmin-connect-sleep-stage_WRSD-PMI-01_c3ea8756.csv'), skiprows=5)

sleep_summary = pd.read_csv(os.path.join('..', 'data','pmi-mini_pilot','4de37ee3-7baf-4592-a431-a6d01fff6bd1',
                                      'WRSD-PMI-01_c3ea8756-8e12-45b7-a31c-1a03e398daf9',
                                      'garmin-connect-sleep-summary',
                                      '000000_garmin-connect-sleep-summary_WRSD-PMI-01_c3ea8756.csv'), skiprows=5)
```

```python
sleep_data['isoDate'] = pd.to_datetime(sleep_data['unixTimestampInMs'] + sleep_data['timezoneOffsetInMs'], unit='ms', utc=True)
sleep_data['isoDate'] = sleep_data['isoDate'].dt.tz_localize(None)

sleep_summary['calendarDate'] = pd.to_datetime(sleep_summary['calendarDate'], format='%Y-%m-%d')
```

```python
calendar_date = datetime.datetime(2023, 3, 23)
```

```python
sleep_stat_row = sleep_summary[sleep_summary.calendarDate == calendar_date]
sleep_start_time = pd.to_datetime((sleep_stat_row['unixTimestampInMs'] + sleep_stat_row['timezoneOffsetInMs']), unit='ms', utc=True).dt.tz_localize(None).iloc[0]
sleep_end_time = pd.to_datetime((sleep_stat_row['unixTimestampInMs'] + sleep_stat_row['timezoneOffsetInMs'] + sleep_stat_row['durationInMs']), unit='ms', utc=True).dt.tz_localize(None).iloc[0]
```

```python
sleep_data[(sleep_data.isoDate >= sleep_start_time) & (sleep_data.isoDate <= sleep_end_time)]
```

```python
intervals = int(divmod((sleep_end_time - sleep_start_time).total_seconds(), 60)[0])
one_minute_intervals = [sleep_start_time + i * datetime.timedelta(minutes=1) for i in range(intervals)]

hypnogram = pd.DataFrame(data={
    'isoDate': one_minute_intervals 
})
```

```python
hypnogram = hypnogram.merge(sleep_data.loc[:, ['isoDate','type']], how='left', on='isoDate')
hypnogram.head()
```

```python
hypnogram['type'] = hypnogram.loc[:, 'type'].fillna(method='ffill')
```

```python
hypnogram
```

```python
"""
The default hypnogram format in YASA is a 1D integer vector where:
-2 = Unscored
-1 = Artefact / Movement
0 = Wake -> awake
1 = N1 sleep -> light
2 = N2 sleep
3 = N3 sleep -> deep
4 = REM sleep -> rem
"""
def from_garmin_to_yasa(x):
    if x == 'rem':
        return 4
    elif x == 'awake':
        return 0
    elif x == 'deep':
        return 3
    else:
        return 1
```

```python
hypnogram['type-yasa'] = hypnogram['type'].apply(from_garmin_to_yasa)
yasa.sleep_statistics(hypnogram['type-yasa'], 1/60)
```
